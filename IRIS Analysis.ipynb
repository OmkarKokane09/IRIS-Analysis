# Label Encoder

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode the 'species' column into numeric values
df['species'] = label_encoder.fit_transform(df['species'])

# Display the updated DataFrame
df.head()
```

# Model Training

```python
from sklearn.model_selection import train_test_split

# Assuming you have previously converted the 'species' column to numeric values
Y = df['species']  # Use 'species' not 'Species'
X = df.drop(columns=['species'])  # Use 'species' not 'Species'

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# model training
model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```
